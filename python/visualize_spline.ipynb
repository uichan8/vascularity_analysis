```python
import numpy as np
import hermite_spline as h
import matplotlib.pyplot as plt

point1 = np.array([2,2])
point2 = np.array([3,4])
grad1 = 2
grad2 = 3

K = 0.005

X = h.hermite_spline(0,point1[0],grad1,1,point2[0],grad2)
Y = h.hermite_spline(0,point1[0],grad1/K,1,point2[0],grad2/K)
Z = h.hermite_spline(0,point1[0],grad1/1000,1,point2[0],grad2/1000)

X_ = h.hermite_spline(0,point1[1],grad1,1,point2[1],grad2)
Y_ = h.hermite_spline(0,point1[1],grad1/K,1,point2[1],grad2/K)
Z_ = h.hermite_spline(0,point1[1],grad1/1000,1,point2[1],grad2/1000)

x = np.linspace(0,1,100)
y1 = h.substitute(X,x)
y2 = h.substitute(Y,x)
y3 = h.substitute(Z,x)

print(X)
print(Y)
print(Z)

y1_ = h.substitute(X_,x)
y2_ = h.substitute(Y_,x)
y3_ = h.substitute(Z_,x)

plt.scatter(x,y1, c = 'r')
plt.scatter(x,y2)
plt.scatter(x,y3, c = 'g')
plt.show()

plt.scatter(x,y1_, c = 'r')
plt.scatter(x,y2_)
plt.scatter(x,y3_, c = 'g')
plt.show()

plt.scatter(y1,y1_, c = 'r')
plt.scatter(y2,y2_)
plt.scatter(y3,y3_, c = 'g')
plt.show()

```

```python
plt.scatter(h.substitute(X,x),h.substitute(X_,x),point1[1],c = 'b')
plt.scatter(h.substitute(Y,x),h.substitute(Y_,x),point2[1],c = 'r')
plt.scatter(h.substitute(Z,x),h.substitute(Z_,x),point2[1],c = 'g')

plt.show()
```
