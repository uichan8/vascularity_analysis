```python
import numpy as np
import hermite_spline as h
import matplotlib.pyplot as plt

point1 = np.array([0,0])
point2 = np.array([1,2])
grad1 = 2
grad2 = 3

K = 1

X = h.hermite_spline(0,point1[0],K,1,point2[0],K)
X_ = h.hermite_spline(0,point1[1],K/grad1,1,point2[1],K/grad2)

x = np.linspace(0,1,100)

y1 = h.substitute(X,x)
y1_ = h.substitute(X_,x)

print(X)
print(X_)


plt.scatter(y1,y1_, c = 'r')
plt.show()

```
